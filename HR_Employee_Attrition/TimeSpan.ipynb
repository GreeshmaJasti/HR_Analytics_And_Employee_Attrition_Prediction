```python
import numpy as np 
import pandas as pd
import pickle
```

```python
data=pd.read_csv('HR.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
for i in data[['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus', 'DailyRate', 'Gender']]:
    data[i]=label_en.fit_transform(data[i])
    print(data[i], end='\n')
```

```python
data['Attrition_ind'] = 0
data.loc[data['Attrition'] == 'Yes', 'Attrition_ind'] = 1
data.head()
```

```python
df1 = data[data['Attrition_ind']==1]
q=df1['YearsAtCompany']
p=df1.drop(['Attrition', 'YearsAtCompany','EmployeeCount', 'EmployeeNumber', 'Over18','OverTime', 'StandardHours','Attrition_ind','BusinessTravel','Department','Education','EducationField','EnvironmentSatisfaction','Gender','JobInvolvement','JobLevel', 'JobRole','JobSatisfaction', 'MaritalStatus', 'RelationshipSatisfaction', 'StockOptionLevel', 'TrainingTimesLastYear'], axis=1)
print(p)
print('\n')
print(q)
```

```python
#split the dataset into train and test
from sklearn.model_selection import train_test_split
p_train,p_test,q_train,q_test = train_test_split(p,q,random_state=42, test_size=0.2)
```

```python
#import gradient boosting regressor
from sklearn.ensemble import GradientBoostingRegressor
#create the instance of the model
gbr=GradientBoostingRegressor(n_estimators=91,max_depth=5, min_samples_split=5, random_state=32)
#train the data
gbr.fit(p_train,q_train)
pickle.dump(gbr ,open('model1.pkl','wb'))
```
